```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(0)
```

To generate 100 floating random numbers with a uniform distribution between 1 and 6:

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.uniform.html#scipy.stats.uniform

```python
from scipy.stats import uniform

r_uniform = uniform.rvs(loc=1, scale=5, size=100)
sns.histplot(r_uniform, kde=False)

print(r_uniform.mean())
plt.show()
```

To generate 100 integer random numbers with a uniform distribution between 1 and 9:

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.randint.html#scipy.stats.randint

```python
from scipy.stats import randint

r_unif_di = randint.rvs(1, 10, loc=0, size=100)
sns.histplot(r_unif_di, kde=False)

print(r_unif_di.mean())
plt.show()
```

To generate 100 floating random numbers with a normal distribution with 0 mean & variance equal to 3:

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html#scipy.stats.norm

```python
from scipy.stats import norm

r_normal = norm.rvs(loc=0, scale=3, size=100)
sns.histplot(r_normal, kde=False)

print(r_normal.mean())
plt.show()
```

To generate 100 floating random numbers with an exponential distribution with mean equal to 1/scale:

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html#scipy.stats.expon

```python
from scipy.stats import expon

r_expon = expon.rvs(scale=3, size=100)
sns.histplot(r_expon, kde=False)
print(r_expon.mean())
plt.show()
```

To generate 100 integer random numbers with a Poisson distribution with a shape parameter corresponding to the mean and the variance:

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.poisson.html#scipy.stats.poisson

```python
from scipy.stats import poisson

r_poisson = poisson.rvs(0.6, size=100)
sns.histplot(r_poisson, kde=False)
print(r_poisson.mean())
plt.show()
```

To generate 100 floating random numbers with a Weibull distribution with c > 0 as the shape parameter (k in the Wikipedia chart).

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.weibull_min.html#scipy.stats.weibull_min

```python
from scipy.stats import weibull_min

r_weibull = weibull_min.rvs(1, size=1000)
sns.histplot(r_weibull, kde=False)
print(r_weibull.mean())
plt.show()
```
